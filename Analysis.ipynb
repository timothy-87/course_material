```python
# Import necessary files
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Univariate analysis
- Analysis one varaible at time
## Bivariate analysis
- relatinship between 2 variables

#### Types of dataset
- Numerical
- Categorical

1. Nominal variable
- No logic between labels {colors, gender, countries}
2. Ordinal variable
- Logic and they are in order {primary, seco, uni, poor, good, average ....}

## Multivariate analysis
- realtionship between more than one vairable

```python
# Reading the dataset
student_data = pd.read_csv("StudentsPerformance.csv")
student_data.head()
```

```python
# Summary of dataset
student_data.describe()
```

```python
# summary for no-numeric
student_data.info()
```

```python
student_data.columns
```

```python
# Univarite analysis
# Categorical feature
gender = student_data['gender']
gender_count = gender.value_counts()

gender_count.plot(kind='bar')
plt.show()
```

```python
# histogram - numerical dataset
plt.hist(student_data['reading score'], edgecolor = "black")
plt.show()
```

```python
# boxplot
plt.boxplot(student_data['math score'])
plt.show()
```

```python
# Bivariate analysis - numerical vs numerical
plt.scatter(student_data['math score'], student_data['reading score'], alpha=0.6)
plt.show()
```

```python
np.corrcoef(student_data['math score'], student_data['reading score'])[0, 1]
```

```python
# caterical vs numerical
student_data.boxplot(column='reading score', by='gender')
plt.grid(True)
```

```python
student_data.info()
```

```python
student_data.columns
```

```python
# Heatmap
import seaborn as sns

num_cols = student_data[['math score', 'reading score','writing score']]

corr_matrix = num_cols.corr()

plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap="OrRd", fmt="0.2f")
plt.show()
```
